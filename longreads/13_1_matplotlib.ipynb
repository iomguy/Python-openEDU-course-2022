# Визаулизация в Matplotlib
В этом лонгриде рассмотрим, как визуализировать данные и строить графики с помощью библиотеки `matplotlib`

```python
# pyplot - наиболее популярная часть библиотеки matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Изначально `matplotlib` создавался для работы с `numpy`-массивами.

В качестве примера, ниже построим график уравнения y = x^2 на интервале от 0 до 10.

Точки на оси x часто задаются методом `np.linspace()`, создающим равномерно расположенные друг от друга точки. Чем больше точек - тем более "плавным" и менее "угловатым" получится график.

Для отображения графика воспользуемся методом `plt.plot()`, который строит кусочно-линейный 2D-график по переданным ему значениям массивов

```python
# возьмём, например, 20 точек
x = np.linspace(0, 10, 20)
y = x**2
```

```python
# строки ниже достаточно для отображения графика
plt.plot(x, y)
```

При написании кода в .py-скрипте (а не Jupyter-тетради или Google Colab) необходимо также явно указать `matplotlib`, что график должен быть отображён, с помощью метода `plt.show()`. 

Т.к. в этой тетради включён режим [`%matplotlib inline`](https://ipython.readthedocs.io/en/stable/interactive/plotting.html), метод `plt.show()` добавлять не обязательно, о чём ниже неоднократно предупредит интерпретатор.

Однако если требуется отобразить несколько графиков, после каждого из них следует применить метод `plt.show()`, иначе графики наложатся на одну и ту же фигуру

```python
plt.plot(x, y)
plt.show()
```


Если в графике будет много элементов (кривых, подписей, легенд), имеет смысл явно объявить для них новую фигуру

```python
fig = plt.figure()

plt.plot(x, y)

fig.show()
```

До тех пор, пока к фигуре не применён метод `show()`, вы можете добавлять к ней новые и менять имеющиеся объекты. Например, ниже добавим к фигуре ещё один график. 

Новым графикам автоматически присваиваются разные цвета, если они не указаны пользователеми самостоятельно

```python
fig = plt.figure()

plt.plot(x, y)
plt.plot(x, y + 50)

fig.show()
```

Цвета, тип линий и маркеров, их толщины и другие свойства могут быть заданы пользователем самостоятельно с помощью аргументов метода `plot()`.

Кратко различные типы линий и маркеров описаны, например, [здесь](https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html)

```python
fig = plt.figure()

plt.plot(x, y)
plt.plot(x, y + 50, color="r", linestyle="--", marker="o")

fig.show()
```

Если присвоить объекты `matplotlib` переменным, их свойства можно не только указывать при создании графиков, но и менять с помощью специальных методов-сеттеров

```python
fig = plt.figure()

# создаём графики
sqares,   = plt.plot(x, y)
sqares_2, = plt.plot(x, y + 50)

# меняем свойства второго
sqares_2.set_color("red")
sqares_2.set_linestyle("--")
sqares_2.set_marker("o")

fig.show()
```

К созданной фигуре также можно самостоятельно "прицепить" один или несколько графиков с помощью метода `add_subplot()`.

Ниже изменим цвет фигуры на зелёный, а фона графика - на красный, и посмотрим, какая область относится к одному, а какая - к другому

Ниже аргумент **(111)** метода `add_subplot()` - это: 1я строка, 1й столбец, 1я (и единственная) ячейка на сетке графиков фигуры.

```python
# создаём фигуру
fig = plt.figure()

# создаём и "цепляем" к ней график
ax = fig.add_subplot(111)

# задаём свойства фигуры и графика
fig.set(facecolor = "pink")
ax.set(facecolor = "lightyellow")

# описываем содержимое графика
ax.plot(x,y)

plt.show()
```

Другими методами можно регулировать множество настроек фигур и графиков

```python

fig = plt.figure()
ax = fig.add_subplot(111) 

fig.set_facecolor("pink")
ax.set_facecolor("lightyellow")

ax.plot(x, y, label="1 график")
ax.plot(x, y+50, label="2 график")

# размеры
ax.set_xlim([-10, 10])
ax.set_ylim([-100, 100])

# подписи
ax.set_title("Название")
ax.set_xlabel("Ось абцис (XAxis)")
ax.set_ylabel("Ось ординат (YAxis)")

# вкл. отображение легенды
ax.legend()

# вкл. отображение сетки
ax.grid()

plt.show()
```

Изменить сразу множество свойств фигуры можно методом `set()`

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y, label="1 график")
ax.plot(x, y+50, label="2 график")

fig.set(facecolor = "pink")
ax.set(facecolor = "lightyellow",
       xlim = [-10, 10],
       ylim = [-100, 100],
       title = "Основы анатомии matplotlib",
       xlabel = "ось абцис (XAxis)",
       ylabel = "ось ординат (YAxis)")

ax.legend()
ax.grid()

plt.show()
```

## Расстановка графиков

Matplotlib позволяет отображать поточечные графики, гистограммы, ящики с усами, скрипичные диаграммы, и множество других типов графиков. Узнать о них подробнее можно [на официальном сайте matplotlib](https://matplotlib.org/stable/plot_types/index.html).

Однако сложные типы графиков удобнее отображать с помощью библиотек `Seaborn` и `Plotly`.

Зато с помощью методов самого `matplotlib` может быть удобно расставлять графики внутри одной фигуры (в том числе построенные с помощью `Seaborn`).

Например, ниже 4 графика отображаются на сетке фигуры 4x4. Обратите внимание на аргументы методов `add_subplot()`

```python
# можно вывести несколько графиков

# а другой магической командой - обратно обычными картинками
# %matplotlib inline

fig = plt.figure()

ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 2, 3)
ax_4 = fig.add_subplot(2, 2, 4)

# ticks задают "риски" на осях
ax_1.set(title = "ax_1", xticks=[], yticks=[])
ax_2.set(title = "ax_2", xticks=[], yticks=[])
ax_3.set(title = "ax_3", xticks=[], yticks=[])
ax_4.set(title = "ax_4", xticks=[], yticks=[])

# отобразим на графиках одну или несколько кривых
ax_1.plot(x, y)

ax_2.plot(x, y)
ax_2.plot(x, -y)

ax_3.plot(x,  y-20)
ax_3.plot(x, -y+20)

plt.show()
```

Каждый график может занимать и большее число клеток. Подробнее о размещении графиков в ячейках одной фигуры можно узнать [здесь](https://matplotlib.org/3.1.0/gallery/subplots_axes_and_figures/subplots_demo.html)

```python
fig = plt.figure()
fig.set(facecolor = "pink")

ax_1 = fig.add_subplot(3, 1, 1)
ax_2 = fig.add_subplot(3, 2, 4)
ax_3 = fig.add_subplot(3, 3, 9)

ax_1.set(title = "ax_1", xticks=[], yticks=[])
ax_2.set(title = "ax_2", xticks=[], yticks=[])
ax_3.set(title = "ax_3", xticks=[], yticks=[])

ax_1.plot(x, y)
ax_2.plot(x, y)
ax_2.plot(x, -y)
ax_3.plot(x,  y-20)
ax_3.plot(x, -y+20)

plt.show()
```
